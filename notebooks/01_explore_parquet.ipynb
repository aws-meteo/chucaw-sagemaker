# ECMWF local parquet quick inspection\n
\n
Failure-first check of required columns and baseline filter.

```python
from pathlib import Path
import pandas as pd

parquet_path = Path("../data/20260409180000-6h-scda-fc.parquet")
if not parquet_path.exists():
    raise FileNotFoundError(f"Parquet file not found: {parquet_path}")
df = pd.read_parquet(parquet_path)
print(f"rows_total={len(df)}")
print(df.columns.tolist())
```

```python
required = {"variable", "isobaricInhPa", "latitude", "longitude", "value"}
missing = sorted(required - set(df.columns))
if missing:
    raise ValueError(f"Missing required columns: {missing}")

filtered = df[(df["variable"] == "t") & (df["isobaricInhPa"] == 1000.0)]
filtered = filtered[["latitude", "longitude", "value"]].dropna()
print(f"rows_after_filter={len(filtered)}")
if filtered.empty:
    raise ValueError("Filtered dataset is empty")
```
